```python
tot_df = tweets_df['date'].dt.date.value_counts().sort_index().reset_index()
tot_df.columns = ['date','Total Count']
```

```python
positive_df = tweets_df.loc[tweets_df['sentiment'] == "['positive']"]
neutral_df = tweets_df.loc[tweets_df['sentiment'] == "['neutral']"]
negative_df = tweets_df.loc[tweets_df['sentiment'] == "['negative']"]
```

```python
positive_df1 = positive_df['date'].dt.date.value_counts().sort_index().reset_index()
positive_df1.columns = ['date','Positive Count']

neutral_df1 = neutral_df['date'].dt.date.value_counts().sort_index().reset_index()
neutral_df1.columns = ['date','Neutral Count']

negative_df1 = negative_df['date'].dt.date.value_counts().sort_index().reset_index()
negative_df1.columns = ['date','Negative Count']
```

```python
positive_df1['date'] = pd.to_datetime(positive_df1.date)
negative_df1['date'] = pd.to_datetime(negative_df1.date)
neutral_df1['date'] = pd.to_datetime(neutral_df1.date)
tot_df['date'] = pd.to_datetime(tot_df.date)
```

```python
tot_df = tot_df.set_index('date') 
positive_df1 = positive_df1.set_index('date') 
neutral_df1 = neutral_df1.set_index('date') 
negative_df1 = negative_df1.set_index('date')
```

```python
merge_df1 = pd.merge(positive_df1,neutral_df1,how='outer',left_index=True,right_index=True)
merge_df2 = pd.merge(merge_df1,negative_df1,how='outer',left_index=True,right_index=True)
merge_df3 = pd.merge(merge_df2,tot_df,how='outer',left_index=True,right_index=True)
```

```python
corr = merge_df3.corr()
corr
```

```python
def calculatePercentage(count,total):
    per = (count/total) * 100
    return per
```

```python
merge_df3['Positive Percentage'] = (merge_df3['Positive Count']/merge_df3['Total Count']) * 100
merge_df3['Neutral Percentage'] = (merge_df3['Neutral Count']/merge_df3['Total Count']) * 100
merge_df3['Negative Percentage'] = (merge_df3['Negative Count']/merge_df3['Total Count']) * 100
```

```python
merge_df3
```

```python
merge_df3['Average Sentiment Score'] = (merge_df3['Positive Count'] * 1 + merge_df3['Neutral Count'] * 0 + merge_df3['Negative Count'] * -1) / merge_df3['Total Count']
```
